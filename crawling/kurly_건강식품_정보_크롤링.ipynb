```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Chrome 설정
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-blink-features=AutomationControlled')
options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36')

driver = webdriver.Chrome(options=options)

div_list = []

# 1~3페이지 반복
for page in range(1, 4):
    url = f'https://www.kurly.com/categories/032?filters=&page={page}&per_page=96&sorted_type=1'
    driver.get(url)

    # 페이지 완전 로드 대기
    time.sleep(3)

    # 스크롤을 내려서 동적 콘텐츠 로드
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight/2);")
    time.sleep(1)

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # 여러 가능한 선택자 시도
    divs = soup.find_all('div', class_='css-mfzoxt css-urqszi css-1l0tq1w')

    # if not divs:
    #     # 다른 클래스명 시도
    #     divs = soup.find_all('div', attrs={'data-testid': 'product-card'})

    # if not divs:
    #     # 상품 링크로 찾기
    #     divs = soup.find_all('a', href=lambda x: x and '/goods/' in x)

    print(f"페이지 {page}: {len(divs)}개 상품 발견")
    div_list.extend(divs)

driver.quit()

print(f"총 {len(div_list)}개 상품")
```

## 이미지 : img_src_list

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Chrome 설정
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--window-size=1920,1080')  # 창 크기 설정 (중요!)

driver = webdriver.Chrome(options=options)

img_src_list = []

# 1~3페이지 반복
for page in range(1, 4):
    driver.get(f'https://www.kurly.com/categories/032?filters=&page={page}&per_page=96&sorted_type=1')

    # 초기 로딩 대기
    time.sleep(3)

    # 천천히 여러 번 스크롤
    screen_height = driver.execute_script("return window.innerHeight")
    i = 1

    while True:
        # 화면 높이만큼씩 천천히 스크롤
        driver.execute_script(f"window.scrollTo(0, {screen_height}*{i});")
        i += 1
        time.sleep(1.5)  # 각 스크롤마다 대기

        # 스크롤 높이 계산
        scroll_height = driver.execute_script("return document.body.scrollHeight")
        if (screen_height) * i > scroll_height:
            break

    # 맨 위로 스크롤하면서 다시 한번 확인
    driver.execute_script("window.scrollTo(0, 0);")
    time.sleep(1)

    # Selenium으로 직접 이미지 요소 찾기 (BeautifulSoup 대신)
    img_elements = driver.find_elements(By.CSS_SELECTOR, 'div.css-jy6mft img')

    for img in img_elements:
        # src, data-src 등 여러 속성 확인
        src = img.get_attribute('src') or img.get_attribute('data-src') or img.get_attribute('data-lazy-src')
        if src and 'placeholder' not in src and 'data:image' not in src:
            if not src.startswith('https://product-image.kurly.com/sticker/image/'):
                img_src_list.append(src)

driver.quit()

print(f"총 이미지 URL 개수: {len(img_src_list)}")
print("첫 5개 이미지 URL:")
for i, url in enumerate(img_src_list[:5], 1):
    print(f"{i}. {url}")
```

### 완성본

```python
import pandas as pd
import re

def separate_name_and_volume(text):
    """
    상품명에서 이름과 분량을 분리하는 함수
    """
    if pd.isna(text):
        return text, None

    # 분량 패턴 정의 (구체적인 것부터 먼저)
    volume_patterns = [
        # 괄호 안의 일수 표현 (60일분), (30일) 등
        r'\([^)]*\d+\s*(?:일분|일|개월분|개월|주분|주)[^)]*\)',
        # 괄호 안의 분량 정보 (45g x 7개입) 등
        r'\([^)]*\d+(?:mg|g|kg|ml|mL|l|L|cc|개|입|포|병|캔|팩|캡슐|정|알)[^)]*\)',
        # 대괄호 안의 분량 정보
        r'\[[^\]]*\d+(?:mg|g|kg|ml|mL|l|L|cc|개|입|포|병|캔|팩|캡슐|정|알)[^\]]*\]',
        # 용량 x 수량 (200mL X 6병, 45g x 7개 등)
        r'\d+(?:\.\d+)?(?:mg|g|kg|ml|mL|l|L|cc|T)\s*[xX×]\s*\d+\s*(?:개|입|포|병|캔|팩|통|박스|매|장)',
        # 단위 + x + 개입 (15mL x 3개입)
        r'\d+(?:\.\d+)?(?:mg|g|kg|ml|mL|l|L|cc|T)\s*[xX×]\s*\d+\s*(?:개입|입|포입|병입|캔입)',
        # 숫자 + 용량 단위 (500ml, 1.5L, 300g 등)
        r'\d+(?:\.\d+)?(?:mg|g|kg|ml|mL|l|L|cc|T)',
        # 수량 표현 (3개입, 10입, 30포 등)
        r'\d+\s*(?:개입|입|포입|병입|캔입|포|병|캔|팩|통|박스|BOX|box)',
        # 캡슐/정/알약 단위
        r'\d+\s*(?:캡슐|정|알|tablet|Tablet|TABLET)',
        # 일수 표현 (괄호 없는 경우)
        r'\d+\s*(?:일분|일|개월분|개월|주분|주)',
    ]

    volumes = []
    text_copy = text

    # 모든 패턴 검색 (이미 매칭된 부분은 제외)
    for pattern in volume_patterns:
        matches = re.finditer(pattern, text_copy)
        for match in matches:
            matched_text = match.group()
            # 이미 추출된 분량과 겹치는지 확인
            is_duplicate = False
            for existing_vol in volumes:
                # 한쪽이 다른 쪽에 포함되면 중복으로 간주
                if matched_text in existing_vol or existing_vol in matched_text:
                    is_duplicate = True
                    break

            if not is_duplicate:
                volumes.append(matched_text)

    if volumes:
        # 분량 정보를 제거하여 깨끗한 상품명 생성
        clean_name = text
        for vol in volumes:
            clean_name = clean_name.replace(vol, '')

        # 여러 공백을 하나로, 앞뒤 공백 제거
        clean_name = re.sub(r'\s+', ' ', clean_name).strip()

        # 남은 특수문자 정리
        clean_name = re.sub(r'\(\s*\)', '', clean_name)
        clean_name = re.sub(r'\[\s*\]', '', clean_name)
        clean_name = re.sub(r'\s+', ' ', clean_name).strip()

        # 분량은 공백으로 구분하여 결합
        volume_str = ' '.join(volumes)

        return clean_name, volume_str
    else:
        return text, None

# 함수 적용
df[['clean_names', 'volumes']] = df['display_name'].apply(
    lambda x: pd.Series(separate_name_and_volume(x))
)
```
